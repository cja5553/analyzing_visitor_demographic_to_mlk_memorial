```python
import numpy as np
import pandas as pd
mlk=pd.read_csv('cbg_data_mlk.csv')
```

```python
visitor_cbg=mlk["visitor_home_cbgs"]
date=mlk["date_range_start"]
```

```python
import json
def string_to_dict(dict_string):
    # Convert to proper json format
    dict_string = dict_string.replace("'", '"')
    return json.loads(dict_string)
```

```python
census_block_list=[]
count_list=[]
month_list=[]
for i in range(len(visitor_cbg)):
  record=visitor_cbg[i]
  record=string_to_dict(record)
  #print(type(record))
  month=date[i]
  for j in record:
    census_block=j
    count=record[j]
    month=month
    census_block_list.append(str(census_block))
    count_list.append(int(count))
    month_list.append(str(month))



```

```python
new_data = pd.DataFrame(list(zip(month_list, census_block_list,count_list)),
               columns =['month', 'home_cbg','count'])
```

```python
import csv
new_data.to_csv("visitor_records.csv")
```
